# Data Generator (Commercial Arrival/Departure)

## Morning Arrival Commercial Feeders

```python
from scipy.stats import genextreme

genloc=8.9068    #Shift
genscale=0.5492  #Scale
genc=0.0629  #Shape

x = np.linspace(genextreme.ppf(0.01, genc,loc=genloc,scale=genscale),
                genextreme.ppf(0.99, genc,loc=genloc,scale=genscale), 100)

genvalmarr = genextreme.rvs(genc, loc = genloc, scale=genscale, size=8761)
plt.figure()
fig, ax = plt.subplots(1, 1)
ax.plot(x, genextreme.pdf(x, genc,loc=genloc,scale=genscale),
       'r-', lw=5, alpha=0.6, label='genextreme pdf')
ax.hist(genvalmarr, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
plt.figure()
plt.plot(genvalmarr)
plt.show()
```

# Data Frame Creation

```python
# Convert the numbers into datetime.time
def number_to_datetime(number):
    if not isinstance(number[-1], datetime.time):
        result=[]
        for val in number:
            temp = str(datetime.timedelta(hours=val))
            result.append(temp)
        del number
        number = pd.to_datetime(result, format='%H:%M:%S.%f').time
    return number

genvalmarr = number_to_datetime(genvalmarr)
genvalevdep = number_to_datetime(genvalevdep)
genvalngdep = number_to_datetime(genvalngdep)
WeibullValEvnArr = number_to_datetime(WeibullValEvnArr)
genvalEvArr = number_to_datetime(genvalEvArr)
genvalRmordep = number_to_datetime(genvalRmordep)
```

```python
df=pd.DataFrame({'Commercial Morning Arrival': genvalmarr,
                 'Commercial Evening Departure': genvalevdep,
                 'Commercial Night Departure': genvalngdep,
                 'Commercial Evening Arrival': WeibullValEvnArr,
                 'Residential Evening Arrival': genvalEvArr,
                 'Residential Morning Departure': genvalRmordep,
                 'Travelled distance': genvalDistance,
                 #'SOCi':Socinitial,  
                })
df.to_excel('EVbehaviour.xlsx')
```

```python
dfv=pd.read_excel('(BEV) Full Discharge 2016 Nissan Leaf 30 kWh UDDS.xlsx',index_col=0)
dfv['distance']=dfv['distance']*1.609 # Convert miles to km
maxEnergy=dfv.essCurKwh[0]
dfv['ChgEnergyReq']=maxEnergy-dfv['essCurKwh']
del dfv['mpsAch']
del dfv['essKwOutAch']
del dfv['speed']
```

```python
# Use the distance from generated data and find closest distance in
# powertrain simulations, merge the resulting DataFrame to the main
# DataFrame to Get SOC on the respective random distance. 

tempind=[]
for inter,x in enumerate(df['Travelled distance']):
    tempind.append(dfv['distance'].sub(x).abs().idxmin())

dfv=dfv.loc[tempind]
dfv=dfv.reset_index()
del dfv['index']

dfmega=df.join(dfv)
#del dfmega['SOCi']
```

# Calculate Departure Times based on Charger Levels

Different Charging levels  SAE J1772_201210 

* level-1  , voltage: 120V AC    ,   Power Level: 1.8KW,
* level-2  , voltage: 208-240V AC ,  Power Level: 19.2 KW,
* level-3  , voltage: 480V DC  ,     Power Level: 50KW-150KW 

```python
dfch= dfmega2
dfch=dfch.drop('Commercial Evening Departure',axis=1)
dfch=dfch.drop('Commercial Night Departure',axis=1)
dfch=dfch.drop('Residential Morning Departure',axis=1)
#dfch

# Calculation of the charging time required to reach 0.95(95%) of SOC 
# Tch=(0.95-SOC)* (CAPACITY OF BATTERY)/(RATED POWER OF CHARGER)* EFFICIENCY OF CHARGER
Tch=((dfch['ChgEnergyReq'])/(1.8)*(0.95)) # Level 1 Charging

Tch2=((dfch['ChgEnergyReq'])/(19.2)*(0.95)) # Level 2 Charging

Tch3=((dfch['ChgEnergyReq'])/(50)*(0.95)) # Level 3 Charging

```

## Commercial Morning Arrival

```python
#Level1 charging 
depT=[]
for at,ct in zip(dfch['Commercial Morning Arrival'].values,Tch):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(1,'Commercial Morning Departure Level 1',depT)
#Level 2 Charging 
depT=[]
for at,ct in zip(dfch['Commercial Morning Arrival'].values,Tch2):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(2,'Commercial Morning Departure Level 2',depT)
#Level 3 charging 
depT=[]
for at,ct in zip(dfch['Commercial Morning Arrival'].values,Tch3):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(3,'Commercial Morning Departure Level 3',depT)
```

## Commercial Evening Arrival

```python
#Level1 charging 
depT=[]
for at,ct in zip(dfch['Commercial Evening Arrival'].values,Tch):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(4,'Commercial Evening Departure Level 1',depT)

#Level 2 Charging 
depT=[]
for at,ct in zip(dfch['Commercial Evening Arrival'].values,Tch2):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(5,'Commercial Evening Departure Level 2',depT)

#Level 3 charging 
depT=[]
for at,ct in zip(dfch['Commercial Evening Arrival'].values,Tch3):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(6,'Commercial Evening Departure Level 3',depT)
```

## Residential Evening Arrival

```python
#Level1 charging 
depT=[]
for at,ct in zip(dfch['Residential Evening Arrival'].values,Tch):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(7,'Residential Evening Departure Level 1',depT)

#Level 2 Charging 
depT=[]
for at,ct in zip(dfch['Residential Evening Arrival'].values,Tch2):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(8,'Residential Evening Departure Level 2',depT)

#Level 3 charging 
depT=[]
for at,ct in zip(dfch['Residential Evening Arrival'].values,Tch3):
    temp=datetime.datetime.combine(datetime.date.today(),at)+datetime.timedelta(hours=ct)
    depT.append(temp.time())
dfch.insert(10,'Residential Departure Level 3',depT)
```
