# Commercial Feeder EV Arrival and Departure statistics 


#Morning Arrival Commercial Feeders

```python
from scipy.stats import genextreme

genloc=8.9068    #Shift
genscale=0.5492  #Scale
genc=0.0629  #Shape

x = np.linspace(genextreme.ppf(0.01, genc,loc=genloc,scale=genscale),
                genextreme.ppf(0.99, genc,loc=genloc,scale=genscale), 100)

genvalmarr = genextreme.rvs(genc, loc = genloc, scale=genscale, size=300)
```

```python
plt.figure()
fig, ax = plt.subplots(1, 1)
ax.plot(x, genextreme.pdf(x, genc,loc=genloc,scale=genscale),
       'r-', lw=5, alpha=0.6, label='genextreme pdf')
ax.hist(genvalmarr, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
plt.figure()
plt.plot(genvalmarr)
plt.show()
```

```python
import datetime

result=[]

for val in genvalmarr:
    temp = str(datetime.timedelta(hours=val))
    result.append(temp)
del genvalmarr
genvalmarr=pd.to_datetime(result, format='%H:%M:%S.%f').time

result=[]
for val in genvalevdep:
    temp = str(datetime.timedelta(hours=val))
    result.append(temp)
del genvalevdep
genvalevdep=pd.to_datetime(result, format='%H:%M:%S.%f').time

result=[]
for val in genvalngdep:
    temp = str(datetime.timedelta(hours=val))
    result.append(temp)
del genvalngdep
genvalngdep=pd.to_datetime(result, format='%H:%M:%S.%f').time

result=[]
for val in WeibullValEvnArr:
    temp = str(datetime.timedelta(hours=val))
    result.append(temp)
del WeibullValEvnArr
WeibullValEvnArr=pd.to_datetime(result, format='%H:%M:%S.%f').time

result=[]
for val in genvalEvArr:
    temp = str(datetime.timedelta(hours=val))
    result.append(temp)
del genvalEvArr
genvalEvArr=pd.to_datetime(result, format='%H:%M:%S.%f').time

result=[]
for val in genvalRmordep:
    temp = str(datetime.timedelta(hours=val))
    result.append(temp)
del genvalRmordep
genvalRmordep=pd.to_datetime(result, format='%H:%M:%S.%f').time
```
